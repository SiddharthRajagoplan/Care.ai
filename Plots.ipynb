```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data from the CSV file
df = pd.read_csv('results.csv')

# Set the style of the plot
sns.set_style("whitegrid")

# Create subplots
plt.figure(figsize=(15, 10))

# Plot 1: Loss vs. Epoch
plt.subplot(2, 2, 1)
sns.lineplot(x='epoch', y='loss', data=df, color='blue', marker='o', markersize=8)
plt.title('Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)

# Plot 2: Learning Rate vs. Epoch
plt.subplot(2, 2, 2)
sns.lineplot(x='epoch', y='learning_rate', data=df, color='green', marker='s', markersize=8)
plt.title('Learning Rate vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.grid(True)

# Plot 3: Step vs. Epoch
plt.subplot(2, 2, 3)
sns.barplot(x='epoch', y='step', data=df, color='purple')
plt.title('Step vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Step')
plt.grid(axis='y')

# Remove empty subplot
plt.delaxes(plt.subplot(2, 2, 4))

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
